## Mencari nilai utility untuk perengkingan

```python
def utility(cmax, cout, cmin):
    return (cmax - cout) / (cmax - cmin)

def utility_dataframe_single(hcmax, hcmin, df, idx):
    cmax = hcmax.iloc[idx] #memilih cmax dari himpunan cmax pada idx
    cmin = hcmin.iloc[idx] #memilih cmin dari himpunan cmin pada idx
    hcout = df.iloc[idx] #memilih himpunan cout pada record tertentu (idx) dari dataframe
    tcout = len(hcout) #hitung jumlah himpunan cout
    for cnum in range(0, tcout):
        #memilih cout dari himpunan cout pada cnum
        cout = hcout[cnum] 
        #mengubah nilai himpunan cout dengan formulasi utility
        hcout[cnum] = utility(cmax, cout, cmin)
    
def utility_count_multiple(hcmax, hcmin, df):
    tdata = len(df) #mendapatkan jumlah record dari dataframe
    for idx in range(0, tdata): 
        # memanggil fungsi utility_dataframe_single
        utility_dataframe_single(hcmax, hcmin, df, idx)


# menjalankan fungsi utility_count_multiple
utility_count_multiple(hcmax, hcmin, df2)
```
